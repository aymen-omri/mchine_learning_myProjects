### importing necessary libraries

```python
import pandas as pd
```

###  read csv file

```python
df = pd.read_csv('./pima-indians-diabetes.data.csv')
df
```

### add columns to my dataframe

```python
df.columns = ["NumTimesPrg","PlGlcConc","BloodP","SkinThick","TwoHourSerIns","BMI","DiPedFunc","age","HasDiabetes"]
df
```

```python
df.notnull()
```

```python
new_df = df['SkinThick'][df.SkinThick != 0]
new_df
```

```python
new_df.median()
```

### replace SkinThick' null attributes with the median

```python
df['SkinThick'][df.SkinThick==0] = new_df.median()
df
```

### Bring the values of all attributes into an interval [0-1]

```python
# MinMaxScaler: MinMaxScaler() put all numeric values on a scale from 0 to 1.
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import MinMaxScaler
 
# Select numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64', 'int32']).columns
print(num_cols)
 
# Impute missing values
for col in num_cols:
    fill_value = df[col].mean()
    df[col].fillna(fill_value, inplace=True)
# Apply MinMaxScaler
minmax = MinMaxScaler()
df[num_cols] = minmax.fit_transform(df[num_cols])
df[num_cols]
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
 
ss = StandardScaler()
df[num_cols] = ss.fit_transform(df[num_cols])
df[num_cols].describe()
```
